```python
import jax

def directional_loss(theta, y, scale=10.0):
    desired_sign = 1.0 - 2.0 * y  # +1 for y=0, -1 for y=1
    violation = theta * desired_sign
    return jax.nn.softplus(-scale * violation) / scale
```
