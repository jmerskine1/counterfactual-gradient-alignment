```python
import os
from dotenv import load_dotenv
from google import genai
import counterfactual_alignment.llm_tools as lt
import pandas as pd

# Load variables from .env file
load_dotenv()

# Access your API key
api_key = os.getenv("GEMINI_API_KEY")

if not api_key:
    raise ValueError("API key not found. Make sure it's set in your .env file.")

# Now you can use the api_key variable securely
client = genai.Client(api_key=api_key)
```

```python
# Load Data

df = pd.read_csv('data/dev_paired.tsv', sep='\t')
```

```python
test_df = df.iloc[0:2]
```

```python
neg = test_df['Text'][test_df['Sentiment']=='Negative'].tolist()
pos = test_df['Text'][test_df['Sentiment']=='Positive'].tolist()
```

```python
response = lt.generate_review(client,pos[0],neg[0])
response2 = lt.generate_review(client,pos[0],response.text)
response3 = lt.generate_review(client,response.text,neg[0])
```
